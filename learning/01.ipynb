## **Typecasting**

Typecasting is a process of converting a one data type yo another.

str(), int(), float(), bool()

```python
# lets create some variables
name = "Prajwal Kamble"
age = 25
gpa = 7.3
is_student = False

print("before typecating:")
print(type(name))
print(type(age))
print(type(gpa))
print(type(is_student))

tc_name = bool(name)
tcs_age = str(age)
tc_age = float(age)
tc_gpa = int(gpa)
print("after typecasting:")
print(type(tc_name),tc_name)
print(type(tcs_age),tcs_age)
print(type(tc_age),tc_age)
print(type(tc_gpa),tc_gpa)

```

## **Arithmetic Operators**
1) (+) => Addition
2) (-) => Substraction
3) (*) => Multiplication
4) (/) => Dividion
5) (//) => Floor Division
6) (%) => Modulus

```python
friends = 0

friends = friends +10
print(friends)

# below are argumented assignment operators
friends -= 2
print(friends)

friends *=5
print(friends)

friends /= 2
print(friends)

friends //=3
print(friends)

friends %=15
print(friends)
```

# âœ… **3. What is shallow copy vs deep copy?**

### **Shallow Copy**

* Copies only outer structure
* Inner objects are shared
* Methods: `copy.copy()`, list slicing, `.copy()`

### **Deep Copy**

* Copies EVERYTHING
* Inner nested objects get new memory
* Method: `copy.deepcopy()`

**Interview line:**

> Shallow copy copies outer objects, deep copy copies nested objects too.

# âœ… **5. Explain memory management in Python.**

Python memory management includes:

### **1. Private Heap Space**

Where all objects live.

### **2. Memory Manager**

Handles allocation & deallocation.

### **3. Reference Counting**

Counts how many references point to an object.

### **4. Garbage Collector**

Removes objects with zero references.

### **5. Pymalloc**

Fast allocator for small objects.

**Interview line:**

> Python uses reference counting and a garbage collector to free unused objects.

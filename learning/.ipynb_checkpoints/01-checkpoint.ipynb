## **Typecasting**

Typecasting is a process of converting a one data type yo another.

str(), int(), float(), bool()

```python
# lets create some variables
name = "Prajwal Kamble"
age = 25
gpa = 7.3
is_student = False

print("before typecating:")
print(type(name))
print(type(age))
print(type(gpa))
print(type(is_student))

tc_name = bool(name)
tcs_age = str(age)
tc_age = float(age)
tc_gpa = int(gpa)
print("after typecasting:")
print(type(tc_name),tc_name)
print(type(tcs_age),tcs_age)
print(type(tc_age),tc_age)
print(type(tc_gpa),tc_gpa)

```

## **Arithmetic Operators**
1) (+) => Addition
2) (-) => Substraction
3) (*) => Multiplication
4) (/) => Dividion
5) (//) => Floor Division
6) (%) => Modulus

```python
friends = 0

friends = friends +10
print(friends)

# below are argumented assignment operators
friends -= 2
print(friends)

friends *=5
print(friends)

friends /= 2
print(friends)

friends //=3
print(friends)

friends %=15
print(friends)
```

```python
# calculate circumferance of circle 2pir
import math
r = float(input("Please Enter radius: "))
pi = math.pi
c = 2 * pi * r
print(f"The circumference of circle is {round(c,4)} cm.")
```

```python
# calculate area of circle
# a = pi*r**2
import math
r = float(input("Enter the radius: "))
r_squre = pow(r,2)
a = math.pi * r_squre
print(f"Area of circle is {round(a,2)} mm^2")
```

```python
# hypoteneous of right angle triangle c = sqrt(a^2 +b^2)
import math
a = float(input(f"Enter a: "))
b = float(input(f"Enter b: "))
print(f"Hypoteneous of right angle triangle is {math.sqrt(pow(a,2)+pow(b,2))}")
```

## **Conditional Statements**

if = do some code only if some condition is True

else do something else

```python
age = int(input("Type your age :"))
if age >= 100:
    print("You are lucky old man")
elif age >=18:
    print("Eligible fore something!")
elif age <= 0:
    print("are you idiot?")

else:
    print("Bye bye you are not eligible")
```
